# Gender vs Degree

```python
gender_degree_dis=df.groupby("degree").gender
gender_degree_dis.value_counts()
```

```python
gender_degree_dis.value_counts().plot(kind="bar",title="Degree vs Gender" ,ylabel="count")
```

```python
highest_earner=df.groupby(["degree","gender"]).income.mean().reset_index().sort_values(by="income")
highest_earner.pivot
```

```python
highest_earner.pivot(index='degree', columns='gender', values='income').plot(kind="bar",ylabel="Max Income",title="Max Income by Degree and Gender")
```
